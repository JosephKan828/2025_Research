## Merge CloudSat data based on Time

### Import packages

```python
import glob
import numpy as np
import netCDF4 as nc

from matplotlib import pyplot as plt
```

### Load data

```python
# file path
path = "/work/b11209013/2024_Research/CloudSat/CloudSat_Interp/"

# file collection
file_list  = glob.glob(path + "*.nc")

with nc.Dataset(file_list[0], "r") as f:
    dims = {
        key: f.variables[key][:]
        for key in f.dimensions.keys()
    }

    qlw = f.variables["qlw"][:]
    qsw = f.variables["qsw"][:]

plt.contourf(np.nanmean(qlw, axis=(0, 2)))
plt.colorbar()
```
